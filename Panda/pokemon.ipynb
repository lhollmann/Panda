```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# File to load
pokemonSheet = "Resources/Pokemon.csv"
```

```python
# Read with Pandas
pokemonDF = pd.read_csv(pokemonSheet, encoding="UTF-8")
pokemonDF.head()
```

```python
# Extract the following columns: "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", and "Speed"

statDF = pokemonDF.loc[:, ["Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", "Speed"]]

statDF.head(30)
```

```python
# Create a dataframe of the average stats for each type of pokemon.
groupedStatsDF = statDF.groupby('Type 1')
#groupedStatsDF.head()
counts = groupedStatsDF['Type 1'].value_counts()
```

```python
avgDF = groupedStatsDF["HP", "Attack", "Sp. Atk", "Sp. Def", "Speed" ].sum()
avgDF.head()
```
